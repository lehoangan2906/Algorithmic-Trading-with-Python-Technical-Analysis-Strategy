# Section 1: Yahoo finance library

```python
import yfinance as yf
```

```python
# Most simple importation
ticker_list = ['GOOG']

yf.download(ticker_list)
```

```python
# Most simple importation
ticker_list = ['GOOG', "EURUSD=X"]

yf.download(ticker_list)
```

```python
# Personalize data range
yf.download(ticker_list, start = '2000-01-01', end = "2010-01-01")
```

```python
# Personalize data interval
yf.download(ticker_list, interval = "1mo")
```
